```python
class Song:
    def __init__(self, titol, autor, durada):
        if (type(titol) != str) or (type(autor) != str) or (type(durada) != (int or float))  :
            raise TypeError
        self.t = titol
        self.a = autor
        self.d = durada
    def info(self):
        return f"{self.t} - {self.a} ({self.d})"
class Playlist:
    def __init__(self, lista):
        if type(lista) != list:
            raise TypeError
        self.lista = lista
    def add(self, song):
        if type(song) != Song:
            raise TypeError
        self.lista.append(song)
    def dtot(self):
        dt = 0
        for song in self.lista:
            dt = dt + song.d
        return dt
    def mostra(self):
        return [a.info() for a in self.lista]
    def cancoll(self):
        cm = None
        cmt = 0
        for song in self.lista:
            d = song.d
            if d > cmt:
                cmt = song.d
                cm = [song.t]
            elif d == cmt and d != 0:
                cm = [a for a in cm] + [song.t]
        return cm

a = Song("mini", "s" , 23)
B = Song("bini", "s" , 40)
print(a.d)
p = Playlist([a, B])
print(p.dtot(), p.cancoll())

```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        if (type(nom) != str) or (type(vida) != int) or (type(forca) != int) or (type(nivell) != int) :
            raise TypeError
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell
    def atacar(self, p2):
        if type(p2) != Personatge:
            raise TypeError
        p2.vida = p2.vida - self.forca
    def pujarnivell(self):
        self.nivell = self.nivell + 1
        self.forca = self.forca + 2
        self.vida = self.vida + 5
    def esta_viu(self):
        if self.vida <= 0:
            return False
        else:
            return True
    def __str__(self):
        return f"nom = {self.nom}, vida = {self.vida}, forca = {self.forca}, nivell = {self.nivell}"
    
P1 = Personatge("Jaume, el conqueridor", 40, 10, 1)
P2 = Personatge("El Cid", 60, 8, 1)

while(P1.esta_viu() == True) and (P2.esta_viu() == True):
    P1.atacar(P2)
    print(f"{P1.nom} ha atacat {P2.nom} amb {P1.forca} de forca")
    P2.atacar(P1)
    print(f"{P2.nom} ha atacat {P1.nom} amb {P2.forca} de forca")
    print("----------------------------------------------------------------------")
if (P1.esta_viu() == True) and (P2.esta_viu() == False):
    print(f"{P1.nom} ha guanyat")
    P1.pujarnivell()
    print(P1.__str__())
elif (P1.esta_viu() == False) and (P2.esta_viu() == True):
    print(f"{P2.nom} ha guanyat")
    P2.pujarnivell()
    print(P2.__str__())
else:
    print("els dos han mort alhora")


```
